```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x_train_df = pd.DataFrame(x_train)
cor = x_train_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r, xticklabels=df.columns[:-1], yticklabels=df.columns[:-1])
plt.title('Correlation Heatmap')
plt.show()
```

```python
column_to_drop = 3 
x_train_cor = np.delete(x_train, column_to_drop, axis=1)
x_test_cor=np.delete(x_test,column_to_drop,axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_cor=sc.fit_transform(x_train_cor)
x_test_cor=sc.fit_transform(x_test_cor)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(x_train_cor,y_train)
```

```python
y_pred = regressor.predict(x_test_cor)
y_pred_reshaped = y_pred.reshape(len(y_pred), 1)
y_test_reshaped = y_test.reshape(len(y_test), 1)
result = np.concatenate((y_pred_reshaped, y_test_reshaped), axis=1)
np.set_printoptions(precision=5)
print(result)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', label='Ideal Line')
plt.title('Linear Regression: Predictions vs. Actual Values')
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```
