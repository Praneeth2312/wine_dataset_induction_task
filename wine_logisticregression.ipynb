```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path='wine.csv'
df=pd.read_csv(file_path,delimiter=';',quotechar='"')
print(df.head(5))
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
df.hist(bins=50,figsize=(12,8))
plt.show()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
num_features_with_quality_3 = np.sum(x == 3)
print(num_features_with_quality_3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x_train_df = pd.DataFrame(x_train)
cor = x_train_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r, xticklabels=df.columns[:-1], yticklabels=df.columns[:-1])
plt.title('Correlation Heatmap')
plt.show()
```

```python
column_to_drop = 3 


x_train_cor = np.delete(x_train, column_to_drop, axis=1)
x_test_cor=np.delete(x_test,column_to_drop,axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_cor=sc.fit_transform(x_train_cor)
x_test_cor=sc.fit_transform(x_test_cor)
```

```python
print(x_train_cor)
```

```python
print(x_test_cor)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train_cor,y_train)
```

```python
y_pred=classifier.predict(x_test_cor)
result=(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
print(result)
```

```python
print(result[:10])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
y_pred_probabilities = classifier.predict_proba(x_test_cor)[:, 1]


plt.scatter(y_test, y_pred_probabilities, color='blue', alpha=0.5)
plt.title('Logistic Regression: Predicted Probabilities vs. Actual Classes')
plt.xlabel('Actual Classes (0 or 1)')
plt.ylabel('Predicted Probabilities')
plt.show()
```
